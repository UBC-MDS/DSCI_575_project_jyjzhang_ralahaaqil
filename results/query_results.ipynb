```python
import pandas as pd
import sys

sys.path.append('../src')

from bm25 import search
from semantic import faiss_search
```

```python
queries = pd.read_csv("../data/queries.csv")
queries
```

## Easy (Keyword-Based)

```python
for query in queries["easy"]:
    print(f"Query: {query}")
    print("BM25:")
    for score, doc in search(query, top_k=5):
        print(f"    {doc.metadata["product"]}")
    print("Semantic:")
    for doc, score in faiss_search(
        query, index_path = "../outputs/faiss_index", top_k=5
    ):
        print(f"    {doc.metadata["product"]}")
    print("")
```
